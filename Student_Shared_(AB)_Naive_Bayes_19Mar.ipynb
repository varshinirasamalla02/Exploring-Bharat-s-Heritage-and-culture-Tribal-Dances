# **Naive Bayes**

```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
y = df['diabetes']
X = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# Model train
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
# Prediction
y_pred = gnb.predict(X_test)
```

```python
# Evaluate
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

# **Multiple Disease**

```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
df.head()
```

```python
df.info(verbose=True)
```

```python
df.prognosis.value_counts()
```

```python
df.describe()
```

```python
df.columns
```

```python
y = df['prognosis']
X = df.drop(['prognosis'],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# Model train
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
# Prediction
y_pred = gnb.predict(X_test)
```

```python
# Evaluate
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
