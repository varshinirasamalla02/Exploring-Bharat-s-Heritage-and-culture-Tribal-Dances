## **Handle Missing Values**

```python
titanic.dropna()
```

```python
titanic.columns
```

```python
df1 = titanic.drop(['cabin', 'embarked', 'boat', 'body', 'home.dest'],axis=1)
```

```python
df1.dropna()
```

```python
df1.describe()
```

```python
df1.age.median()
```

```python
df1.age.fillna(df1.age.median(),inplace=True)
```

```python
df1.age
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='median')
```

```python
df1[['age_imp','fare_imp']] = si.fit_transform(df1[['age','fare']])
```
